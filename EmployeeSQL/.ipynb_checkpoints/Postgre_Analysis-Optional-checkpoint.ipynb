```python
averages = 'SELECT t.title, avg(s.salary) FROM employee e\
    JOIN employee_salary as s \
ON  (e.emp_no = s.emp_no)\
JOIN titles as t \
ON(t.title_id = e.emp_title_id)\
GROUP BY t.title'
df2 = pd.read_sql_query(averages, connection)
```

```python
df2 = df2.set_index('title')
df2.plot(kind= 'bar')
```
